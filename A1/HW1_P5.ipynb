```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('ngaussian.csv', delimiter=' ', names=['X1', 'X2'])
```

### Part A

Structural Causal Model 1:
$$X_1 = \epsilon_1$$
$$X_2 = \beta_{12} X_1 + \epsilon_2$$

```python
beta_12 = np.polyfit(df['X1'], df['X2'], 1)[0]
print(f'beta_12 = {beta_12:.4f}')
```

```python
residual = df['X2'] - beta_12 * df['X1']

plt.scatter(df['X1'], residual)
plt.xlabel('X1')
plt.ylabel(r'$X2 - \beta_{12} \cdot X1$')
plt.show()
```

### Part B

Structural Causal Model 2:
$$X_2 = \epsilon_2$$
$$X_1 = \beta_{21} X_2 + \epsilon_1$$

```python
beta_21 = np.polyfit(df['X2'], df['X1'], 1)[0]
print(f'beta_21 = {beta_21:.4f}')
```

```python
residual = df['X1'] - beta_21 * df['X2']

plt.scatter(df['X2'], residual)
plt.xlabel('X2')
plt.ylabel(r'$X1 - \beta_{21} \cdot X2$')
plt.show()
```

### Part C

We observe that under the assumption that the data points are generated by Structural Causal Model 1, we have $\epsilon_1 \bot \epsilon_2$, indicating statistical independence. However, when assuming Model 2, we reach the conclusion that $\epsilon_1 \not\bot \epsilon_2$, signifying a lack of independence. This suggests that the first model holds true, supporting the claim that $X_1$ is the cause of $X_2$.
